```python
t = np.arange(1,sr*dur+1)/sr #sampling rate 만의 time tick을 만드는 것. (1초에) >제일 첫번째 time tick 부터 1초까지 완성된 것.
```

```python
theta = t* 2*np.pi *freq  #time을 theta에다 연결시켜 놓음. = 1초에다가 한바퀴의 sine function 입힌 것. frequency는 몇번 반복할지를 적어주는 것.
```

```python
#소리를 만들때에는 time 이 중요하니 x축에는 time을 적는다. 
fig = plt.figure()
ax= fig. add_subplot(111)
ax. plot(t[0:1000], s[0:1000], '.')
ax. set_xlabel('time (s)')
ax. set_ylabel('real')
```
